```python
import pandas as pd
from scenarios import ATTRIBUTES, PEOPLE_IDS
from scenarios import will, alex, sophia, chintan, vin
```

### Define a Scenario

```python
views_in_bulk = []
```

```python
will_dots_alex_negatively = [i for i in will.generate_negative_views(alex, number_of_observations=12)]
will_dots_alex_positively = [i for i in will.generate_positive_views(alex, number_of_observations=3)]
```

```python
alex_dots_will_negatively = [i for i in alex.generate_negative_views(will, number_of_observations=6)]
alex_dots_will_positively = [i for i in alex.generate_positive_views(will, number_of_observations=1)]
```

```python
views_in_bulk.extend(will_dots_alex_negatively)
views_in_bulk.extend(will_dots_alex_positively)
views_in_bulk.extend(alex_dots_will_negatively)
views_in_bulk.extend(alex_dots_will_positively)
```

### Save to DataFrame

```python
scenario = pd.DataFrame(views_in_bulk)
```

```python
attribute_lookup = {v: k for k, v in ATTRIBUTES.items()}
people_lookup = {v: k for k, v in PEOPLE_IDS.items()}
```

```python
scenario['attribute_name'] = scenario['attribute'].apply(lambda x: attribute_lookup.get(x))
scenario['subject_name'] = scenario['subject'].apply(lambda x: people_lookup.get(x))
scenario['author_name'] = scenario['author'].apply(lambda x: people_lookup.get(x))
```
